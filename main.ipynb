```python
def get_artist_id(artist_metadata: List[Dict]):
    return get_artist_id_helper(artist_metadata.copy(), [])

def get_artist_id_helper(artist_metadata: List[Dict], acc: List):
    if len(artist_metadata) == 0:
        return acc
    data = artist_metadata.pop()
    acc.append(data["id"])
    return get_artist_id_helper(artist_metadata, acc)
```

```python
def get_artist_top_track(artist_id: str) -> List[str]:
    tracks = pd.DataFrame(spotify.artist_top_tracks(artist_id, country="TH"))["tracks"].apply(lambda x: pd.Series(x))
    if "id" in tracks:
        return pd.DataFrame(spotify.artist_top_tracks(artist_id, country="TH"))["tracks"].apply(lambda x: pd.Series(x))["id"].tolist()
    return []
```

```python
artist_top_tracks = {}
for a_id in artist_list:
    artist_top_tracks[a_id] = get_artist_top_track(a_id)
```

```python
def flatten(l: List[List]):
    return [item for sublist in l for item in sublist]
```

```python
playlist_metadata["artist_top_tracks"] = playlist_metadata["artist_id"].apply(lambda x: flatten(list(map(lambda y: artist_top_tracks.get(y), x))))
```
